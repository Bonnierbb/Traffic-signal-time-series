```python
# 120 st and Amsterdam Ave
# The red signal is lasted for 50 seconds and green signal is 41 seconds

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

piecewise_time = [0, 50, 65, 75, 90]
piecewise_velocity = [0.000, 0.000, 0.010, 0.010, 0.000]

# Plotting
plt.figure(figsize=(15, 10))

plt.plot(piecewise_time, piecewise_velocity, marker='o', linestyle='-', linewidth=2, color='black', label='Piecewise Line')

plt.title(f'Velocity over Time (Overlapping every {90} seconds)')
plt.xlabel(f'Time Modulo {90} (seconds)')
plt.ylabel('Velocity (km/s)')
plt.xlim(0, 95)  # Set x-axis limit to period_T seconds
plt.grid(True)
plt.legend()
plt.show()
```
